```python
import pandas as pd
import numpy as np

```

```python
on_time_linearized = np.load(f'outputs/x_train_on_linearized.npy')
on_time_linearized = on_time_linearized[:1000,:]
on_time_linearized.shape
```

```python
off_time_linearized = np.load(f'outputs/x_train_off_linearized.npy')
off_time_linearized = off_time_linearized[:1000,:]
off_time_linearized.shape
```

```python
df = pd.DataFrame(columns=list(range(1,17))+['on', 'off', 'noise', '???', 'probe_flag'], index=range(off_time_linearized.shape[0]*off_time_linearized.shape[1]))
df
```

```python
for i in range(1,17):
    df.loc[:,i] = True
df.loc[:,'probe_flag'] = False
df.loc[:,'???'] = 50
df.loc[:,'noise'] = 0
df
```

```python
for i, off_sequence in enumerate(off_time_linearized):
    df.loc[range(i*off_time_linearized.shape[1],(i+1)*off_time_linearized.shape[1]),'off'] = off_sequence
```

```python
for i, on_sequence in enumerate(on_time_linearized):
    df.loc[range(i*on_time_linearized.shape[1],(i+1)*on_time_linearized.shape[1]),'on'] = on_sequence
```

```python
df.isna().sum()
```

```python
for i, on_sequence in enumerate(on_time_linearized):
    np.savetxt(f'outputs/patterns/linearized/{i}.txt', df.values[range(i*on_time_linearized.shape[1],(i+1)*on_time_linearized.shape[1]), :], fmt='%s', delimiter=', ')
    if i ==100:
        break
```

```python
df['on']=df['on'].astype(int)
df['off']=df['off'].astype(int)
```

```python
for i, on_sequence in enumerate(on_time_linearized):
    np.savetxt(f'outputs/patterns/linearized_int/{i}.txt', df.values[range(i*on_time_linearized.shape[1],(i+1)*on_time_linearized.shape[1]), :], fmt='%s', delimiter=', ')
    if i ==100:
        break
```
